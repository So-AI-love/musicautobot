```
import librosa
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
from music21 import stream, note, pitch
import tensorflow as tf
import tensorflow_addons as tfa
from transformers import BasicPitchProcessor, BasicPitchModel

# Function to load an audio file
def load_audio_file(filepath):
    y, sr = librosa.load(filepath, sr=16000)  # Ensures a consistent sample rate
    return y, sr

# Function to extract pitch using Basic Pitch
def extract_pitch_basic_pitch(y, sr):
    processor = BasicPitchProcessor.from_pretrained("spotify/basic-pitch")
    model = BasicPitchModel.from_pretrained("spotify/basic-pitch")

    inputs = processor(y, return_tensors="pt", sampling_rate=sr)
    with torch.no_grad():
        outputs = model(**inputs)

    # Convert the output to MIDI format (or just extract pitches)
    midi_data = processor.decode(outputs.logits[0])
    return midi_data

# Function to extract pitch using TensorFlow SPICE model
def extract_pitch_tf_spice(y, sr):
    y = tf.convert_to_tensor(y, dtype=tf.float32)
    # SPICE model expects frames with a length of 2048
    frame_size = 2048
    frames = tf.signal.frame(y, frame_size, frame_size // 2)
    spice_model = tfa.layers.SPICE(frames_per_hop=512)

    pitch_values = spice_model(frames).numpy()

    # Filter out unvoiced frames
    pitch_values = pitch_values[pitch_values > 0]

    return pitch_values

# Function to map pitch values to musical notes using music21
def map_pitch_to_notes(pitch_values):
    notes = []
    for p in pitch_values:
        if p > 0:
            n = pitch.Pitch()
            n.frequency = p
            notes.append(note.Note(n))
    return notes

# Function to create a music21 stream and display the notes
def create_music_score(notes):
    score = stream.Stream()
    for n in notes:
        score.append(n)

    # Display the score
    score.show()

# Upload the audio file
from google.colab import files
uploaded = files.upload()

# Load and analyze the audio file
for filename in uploaded.keys():
    y, sr = load_audio_file(filename)

    # Choose method: Basic Pitch or TensorFlow SPICE
    method = "basic_pitch"  # Change to "tf_spice" for TensorFlow SPICE method

    if method == "basic_pitch":
        midi_data = extract_pitch_basic_pitch(y, sr)
        # Convert MIDI to notes (if needed)
        notes = map_pitch_to_notes(midi_data["midi"])
    else:
        pitch_values = extract_pitch_tf_spice(y, sr)
        notes = map_pitch_to_notes(pitch_values)

    create_music_score(notes)
```

```
# Install specific versions of TensorFlow and Keras
!pip install tensorflow>=2.13.0,<=2.15.0 keras==2.13.1 keras_applications==1.0.8

# Import necessary libraries
import librosa
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
from music21 import stream, note, pitch, meter, tempo, key, instrument
from collections import Counter
from mpl_toolkits.mplot3d import Axes3D

# Function to load an audio file
def load_audio_file(filepath):
    y, sr = librosa.load(filepath, sr=16000)  # Ensures a consistent sample rate
    return y, sr

# Function to extract pitch using librosa
def extract_pitch(y, sr):
    pitches, magnitudes = librosa.core.piptrack(y=y, sr=sr, fmin=75, fmax=1000)

    pitch_values = []
    for t in range(magnitudes.shape[1]):
        index = magnitudes[:, t].argmax()
        pitch_val = pitches[index, t]
        if pitch_val > 0:  # Consider only positive pitch values
            pitch_values.append(pitch_val)

    # Smooth pitch contour with a Gaussian filter
    pitch_values = gaussian_filter1d(pitch_values, sigma=2)

    return pitch_values

# Function to map pitch values to musical notes using music21
def map_pitch_to_notes(pitch_values):
    notes = []
    for p in pitch_values:
        if p > 0:
            n = pitch.Pitch()
            n.frequency = p
            notes.append(n.nameWithOctave)  # Store the note name instead of the Note object
    return notes

# Function to create a music21 stream and save the notes artistically
def create_music_score(notes):
    score = stream.Score()
    part = stream.Part()

    # Add a key signature, time signature, and tempo
    score.append(key.KeySignature(0))  # C major
    score.append(meter.TimeSignature('4/4'))
    score.append(tempo.MetronomeMark(number=120))  # 120 BPM
    score.append(instrument.Instrument('Piano'))  # Instrument type

    for n in notes:
        if n:  # Ensure the note is valid
            part.append(note.Note(n))  # Append Note objects to the part

    score.insert(0, part)

    # Save the score as a MusicXML file
    score.write('musicxml', fp='output_score.xml')
    print("Score saved as 'output_score.xml'. You can download it from the file system.")

# Function to create a 3D visualization of the detected notes and chords
def visualize_chord_3d(pitch_values, notes):
    # Get the frequencies for the musical notes
    note_freqs = get_note_frequencies()

    # Create a 3D scatter plot
    fig = plt.figure(figsize=(14, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Prepare data for the 3D plot
    time_frames = np.arange(len(pitch_values))

    # Create a list to hold the frequencies of the detected notes
    note_frequencies = []
    for n in notes:
        if n in note_freqs:
            note_frequencies.append(note_freqs[n])
        else:
            note_frequencies.append(0)  # Use 0 for notes not in the defined scale

    # Scatter plot of pitch values over time
    ax.scatter(time_frames, pitch_values, note_frequencies, c='b', marker='o')

    # Annotate each note in the chord
    for i, n in enumerate(notes):
        if n in note_freqs:
            ax.text(time_frames[i], pitch_values[i], note_freqs[n], n, fontsize=10)

    # Set labels
    ax.set_xlabel('Time (frames)')
    ax.set_ylabel('Pitch (Hz)')
    ax.set_zlabel('Note Frequency (Hz)')
    ax.set_title('3D Visualization of Detected Musical Chords')

    plt.show()

# Function to get frequencies for the musical notes (Do, Re, Mi, Fa, Sol, La, Ti)
def get_note_frequencies():
    # Define the frequencies for the notes in the C major scale (Do, Re, Mi, Fa, Sol, La, Ti)
    note_freqs = {
        'Do': 261.63,  # C4
        'Re': 293.66,  # D4
        'Mi': 329.63,  # E4
        'Fa': 349.23,  # F4
        'Sol': 392.00,  # G4
        'La': 440.00,  # A4
        'Ti': 493.88   # B4
    }
    return note_freqs

# Upload the audio file
from google.colab import files
uploaded = files.upload()

# Load and analyze the audio file
for filename in uploaded.keys():
    y, sr = load_audio_file(filename)
    pitch_values = extract_pitch(y, sr)
    notes = map_pitch_to_notes(pitch_values)
    create_music_score(notes)  # Create the score artistically
    visualize_chord_3d(pitch_values, notes)  # Call the 3D chord visualization function
```
